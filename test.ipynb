```python
data = {
  "image_b64": ""
}
```

```python
import requests
import json

url = 'http://localhost:8000/diffusion/outpaint/'  # Flask 애플리케이션의 URL로 변경
headers = {'Content-Type': 'application/json'}  # JSON 형식의 데이터를 보낼 것임

# JSON 데이터를 문자열로 변환하여 전송
response = requests.post(url, headers=headers, data=json.dumps({"body":data}))

# 응답 확인
print('Status Code:', response.status_code)
print('Response Content:', response.text)
```
