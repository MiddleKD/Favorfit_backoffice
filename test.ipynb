## Remove bg 
- not queue

```python
import requests
import json

url = f"{root_endpoint}/utils/remove_bg/"  # Flask  URL 
headers = {'Content-Type': 'application/json'}  # JSON    

remove_bg_body = {"image_b64":image_b64, "post_process":True, "box":[0,0,200,200]}
# JSON    
response = requests.post(url, headers=headers, data=json.dumps({"body":remove_bg_body}))

#  
print('Status Code:', response.status_code)
print('Response Content:', response.text)

mask_b64 = json.loads(json.loads(response.text)["body"])["image_b64"].split("base64,")[1]
```
